```python
# Defeing a function that takes data from sns and put it in the same name parameter
def assigned_dataset():
  # taking dataset name
  dataset_name = sns.get_dataset_names()
  # using for loop to iterate each name with data from sns
  for name in dataset_name:
    # using global variable
    globals()[name] = sns.load_dataset(name)
# printing out only name
assigned_dataset()
```
